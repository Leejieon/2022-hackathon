````python
sales1 = PATH + "대구시뉴스기사크롤링_ver05.csv"
sales2 = PATH + "대구시output.csv"
````

# **1. DataFrame 고치기**



> **1.1 DataFrame 합치기**
* sales1 내용
```
썸네일, 보도일자, 언론사, 기사제목, URL, 기사내용
```
* sales2 내용
```
 카테고리
```




````python
import pandas as pd
# merge files
dataFrame = pd.concat(
   map(pd.read_csv, [sales1, sales2]), ignore_index=True, axis=1)
````

````python
dataFrame
````



> **1.2 불필요한 열 제거**
* 0, 7번 열과 같이 불필요한 index를 나타내는 열은 제거한다.



````python
df = dataFrame.drop([0, 7], axis = 1)
````

````python
df
````



> **1.3 DataFrame 저장하기**


````python
df.to_csv(PATH + '대구시result.csv',index = False)
````



> **1.4 column 이름 설정하기**


````python
cat = ["IT과학", "경제", "사회", "생활문화", "세계", "스포츠", "정치"]

df.columns = ["썸네일", "보도 일자", "언론사", "기사 제목", "URL", "기사내용", "카테고리"]
````

````python
df
````



> **1.5 카테고리 수정하기**
* 기존의 카테고리는 번호로 되어있다. 기존에 mapping된 단어들로 교체하여 준다.

````python
df = df.replace(0, 'IT과학')
df = df.replace(1, '경제')
df = df.replace(2, '사회')
df = df.replace(3, '생활문화')
df = df.replace(4, '세계')
df = df.replace(5, '스포츠')
df = df.replace(6, '정치')
````

````python
df
````



> **1.6 카테고리 수정하기(코로나)**
* 현재 뉴스가 많이 나오는 정보인 코로나 관련 정보는 따로 분류하여 준다.
* 기존의 카테고리가 존재하지만 코로나 관련 정보는 따로 분류하여 주는 것이 뉴스를 보는 유저의 입장에서 더 효율적이라고 판단했다.


````python
df.loc[df['기사 제목'].str.contains('코로나'), '카테고리'] = '코로나'
df.loc[df['기사 제목'].str.contains('확진'), '카테고리'] = '코로나'
df
````

````python
df.to_csv(PATH + '대구시result_corona.csv',index = False)
````
